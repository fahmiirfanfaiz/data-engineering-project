## Tahap 1: Data Cleaning dan Analisis

Rekap harian kasus COVID-19 per kelurahan di Provinsi DKI Jakarta - Mei 2020

````python
# Import library yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from datetime import datetime
import os

# Konfigurasi visualisasi
plt.style.use('default')
sns.set_palette("husl")
warnings.filterwarnings('ignore')

# Konfigurasi pandas untuk menampilkan lebih banyak kolom
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

print("Library berhasil diimport!")
print(f"Pandas version: {pd.__version__}")
print(f"Numpy version: {np.__version__}")
````

````python
# Memuat dataset
data_path = "../data/raw/data-rekap-harian-kasus-covid19-per-kelurahan-di-provinsi-dki-jakarta-bulan-mei-2020.csv"

print("Memuat dataset...")
df = pd.read_csv(data_path)

print(f"Dataset berhasil dimuat!")
print(f"Shape dataset: {df.shape}")
print(f"Jumlah baris: {df.shape[0]}")
print(f"Jumlah kolom: {df.shape[1]}")
````

## 1. Exploratory Data Analysis (EDA)
### Menganalisis struktur dan karakteristik dataset

````python
# Menampilkan informasi dasar dataset
print("=== INFORMASI DASAR DATASET ===")
print("\n1. Info Dataset:")
df.info()

print("\n2. Nama Kolom:")
for i, col in enumerate(df.columns, 1):
    print(f"{i:2d}. {col}")

print("\n3. Preview 5 baris pertama:")
df.head()
````

````python
# Analisis Missing Values
print("=== ANALISIS MISSING VALUES ===")

# Hitung missing values per kolom
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

# Buat DataFrame untuk analisis missing values
missing_analysis = pd.DataFrame({
    'Kolom': missing_values.index,
    'Jumlah_Missing': missing_values.values,
    'Persentase_Missing': missing_percentage.values
}).sort_values('Jumlah_Missing', ascending=False)

print("Missing Values per Kolom:")
print(missing_analysis)

# Total missing values
total_missing = missing_values.sum()
total_cells = np.prod(df.shape)
print(f"\nTotal missing values: {total_missing}")
print(f"Total cells: {total_cells}")
print(f"Persentase missing dari seluruh dataset: {(total_missing/total_cells)*100:.2f}%")
````

````python
# Analisis Duplikasi Data
print("\n=== ANALISIS DUPLIKASI DATA ===")

# Cek duplikasi lengkap
duplicate_rows = df.duplicated().sum()
print(f"Jumlah baris duplikat lengkap: {duplicate_rows}")

# Cek duplikasi berdasarkan key columns (yang meaningful)
key_columns = ['periode_data', 'id_kel', 'nama_kelurahan', 'nama_kecamatan']
duplicate_keys = df.duplicated(subset=key_columns).sum()
print(f"Duplikasi berdasarkan key columns {key_columns}: {duplicate_keys}")

# Analisis lebih detail untuk kolom-kolom penting
if duplicate_keys > 0:
    print("\nBaris dengan key duplikat:")
    duplicated_data = df[df.duplicated(subset=key_columns, keep=False)].sort_values(key_columns)
    print(duplicated_data[key_columns + ['positif', 'sembuh', 'meninggal']])
else:
    print("Tidak ada duplikasi pada key columns - data sudah unik per kelurahan per periode")
````

````python
# Analisis Data Noisy dan Outliers
print("=== ANALISIS DATA NOISY DAN OUTLIERS ===")

# Fokus pada kolom numerik yang berisi data COVID-19
covid_columns = ['odp', 'proses_pemantauan', 'selesai_pemantauan', 'pdp', 
                'masih_dirawat', 'pulang_dan_sehat', 'positif', 'dirawat', 
                'sembuh', 'meninggal', 'self_isolation']

# Statistik deskriptif
print("Statistik Deskriptif untuk kolom COVID-19:")
stats_summary = df[covid_columns].describe()
print(stats_summary)

# Deteksi outliers menggunakan IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

print("\n=== DETEKSI OUTLIERS (IQR Method) ===")
outliers_summary = {}

for col in covid_columns:
    outliers, lower, upper = detect_outliers_iqr(df, col)
    outliers_summary[col] = {
        'count': len(outliers),
        'percentage': (len(outliers)/len(df))*100,
        'lower_bound': lower,
        'upper_bound': upper
    }
    
    if len(outliers) > 0:
        print(f"\n{col}:")
        print(f"  - Jumlah outliers: {len(outliers)}")
        print(f"  - Persentase: {(len(outliers)/len(df))*100:.2f}%")
        print(f"  - Batas bawah: {lower:.2f}")
        print(f"  - Batas atas: {upper:.2f}")
        print(f"  - Nilai outliers: {sorted(outliers[col].unique())}")
    else:
        print(f"\n{col}: Tidak ada outliers")
````

## 3. Implementasi Data Cleaning

````python
# Implementasi Data Cleaning
print("=== DATA CLEANING IMPLEMENTATION ===")

# Buat copy dataset untuk cleaning
df_original = df.copy()
df_cleaned = df.copy()

print("1. MENGHAPUS KOLOM KETERANGAN")
print("   Alasan: 100% missing values, tidak memberikan informasi")

# Drop kolom keterangan
df_cleaned = df_cleaned.drop('keterangan', axis=1)

print(f"   ‚úì Kolom 'keterangan' berhasil dihapus")
print(f"   ‚úì Shape sebelum: {df_original.shape}")
print(f"   ‚úì Shape setelah: {df_cleaned.shape}")

print("\n2. VALIDASI TIDAK ADA MISSING VALUES LAGI")
remaining_missing = df_cleaned.isnull().sum().sum()
print(f"   ‚úì Total missing values setelah cleaning: {remaining_missing}")

print("\n3. ANALISIS OUTLIERS - PENDEKATAN KONSERVATIF")
print("   Untuk data COVID-19, outliers bisa jadi informasi penting (hotspot/cluster)")
print("   Pendekatan: RETAIN outliers dengan flagging untuk analisis lanjutan")

# Buat kolom flag outliers
outlier_flags = pd.DataFrame(index=df_cleaned.index)

for col in covid_columns:
    outliers, lower, upper = detect_outliers_iqr(df_cleaned, col)
    flag_column = f'{col}_outlier_flag'
    outlier_flags[flag_column] = ((df_cleaned[col] < lower) | (df_cleaned[col] > upper)).astype(int)

# Tambahkan summary outlier flag
outlier_flags['total_outlier_flags'] = outlier_flags.sum(axis=1)

# Gabungkan dengan dataset
df_cleaned = pd.concat([df_cleaned, outlier_flags], axis=1)

print(f"   ‚úì Ditambahkan {len(outlier_flags.columns)} kolom flag outliers")
print(f"   ‚úì Baris dengan outliers dalam 1+ kolom: {(outlier_flags['total_outlier_flags'] > 0).sum()}")

print("\n4. VALIDASI KONSISTENSI DATA")
# Cek konsistensi logis: misal total kasus vs komponen
print("   Mengecek konsistensi data COVID-19...")

# Contoh validasi: positif = dirawat + sembuh + meninggal + self_isolation
df_cleaned['calculated_positif'] = (df_cleaned['dirawat'] + df_cleaned['sembuh'] + 
                                   df_cleaned['meninggal'] + df_cleaned['self_isolation'])

inconsistent_positif = df_cleaned[df_cleaned['positif'] != df_cleaned['calculated_positif']]

print(f"   ‚úì Baris dengan inkonsistensi data positif: {len(inconsistent_positif)}")
if len(inconsistent_positif) > 0:
    print(f"     Persentase: {(len(inconsistent_positif)/len(df_cleaned))*100:.2f}%")

print("\n5. RINGKASAN HASIL CLEANING")
print(f"   ‚úì Dataset original: {df_original.shape}")
print(f"   ‚úì Dataset cleaned: {df_cleaned.shape}")
print(f"   ‚úì Kolom dihapus: 1 (keterangan)")
print(f"   ‚úì Kolom ditambah: {len(outlier_flags.columns) + 1} (flags + calculated_positif)")
print(f"   ‚úì Missing values: {df_original.isnull().sum().sum()} ‚Üí {df_cleaned.isnull().sum().sum()}")
print(f"   ‚úì Data integrity: Outliers diflag, konsistensi dicek")
````

## 1. Correlation Analysis - Identifikasi Redundansi Data

````python
# Correlation Analysis untuk Identifikasi Redundansi
print("=== CORRELATION ANALYSIS ===")

# Hitung correlation matrix
correlation_matrix = df_integration[covid_numeric_cols].corr()

# Identifikasi korelasi tinggi (threshold > 0.7 atau < -0.7)
high_correlation_threshold = 0.7
print(f"Threshold korelasi tinggi: |r| > {high_correlation_threshold}")

# Cari pasangan variabel dengan korelasi tinggi
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_val = correlation_matrix.iloc[i, j]
        if abs(corr_val) > high_correlation_threshold:
            var1 = correlation_matrix.columns[i]
            var2 = correlation_matrix.columns[j]
            high_corr_pairs.append((var1, var2, corr_val))

print(f"\nPasangan variabel dengan korelasi tinggi (|r| > {high_correlation_threshold}):")
if high_corr_pairs:
    for var1, var2, corr in high_corr_pairs:
        print(f"  {var1:20s} ‚Üî {var2:20s}: r = {corr:6.3f}")
else:
    print("  Tidak ada korelasi tinggi yang ditemukan")

# Visualisasi Correlation Matrix
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# 1. Heatmap korelasi lengkap
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='RdBu_r', center=0,
            square=True, ax=axes[0], cbar_kws={"shrink": .8})
axes[0].set_title('Correlation Matrix - COVID-19 Variables')

# 2. Heatmap korelasi dengan threshold
correlation_filtered = correlation_matrix.copy()
correlation_filtered[abs(correlation_filtered) < 0.3] = 0  # Filter korelasi rendah
sns.heatmap(correlation_filtered, mask=mask, annot=True, cmap='RdBu_r', center=0,
            square=True, ax=axes[1], cbar_kws={"shrink": .8})
axes[1].set_title('Correlation Matrix - Filtered (|r| > 0.3)')

plt.tight_layout()
plt.show()

# Analisis korelasi berdasarkan kategori
print("\n=== ANALISIS KORELASI BERDASARKAN KATEGORI ===")

# Kelompokkan variabel berdasarkan kategori epidemiologi
categories = {
    'Suspek': ['odp', 'proses_pemantauan', 'selesai_pemantauan'],
    'Probable': ['pdp', 'masih_dirawat', 'pulang_dan_sehat'],
    'Confirmed': ['positif', 'dirawat', 'sembuh', 'meninggal', 'self_isolation']
}

for category, variables in categories.items():
    print(f"\nKategori {category}:")
    if len(variables) > 1:
        cat_corr = correlation_matrix.loc[variables, variables]
        # Ambil korelasi yang signifikan
        for i in range(len(variables)):
            for j in range(i+1, len(variables)):
                corr_val = cat_corr.iloc[i, j]
                if abs(corr_val) > 0.5:  # Threshold lebih rendah untuk dalam kategori
                    print(f"  {variables[i]:20s} ‚Üî {variables[j]:20s}: r = {corr_val:6.3f}")

# Summary correlation statistics
print(f"\n=== SUMMARY CORRELATION ANALYSIS ===")
print(f"‚úì Total variabel: {len(covid_numeric_cols)}")
print(f"‚úì Korelasi tinggi (|r| > {high_correlation_threshold}): {len(high_corr_pairs)} pasangan")
print(f"‚úì Range korelasi: {correlation_matrix.values[~np.eye(len(correlation_matrix), dtype=bool)].min():.3f} - {correlation_matrix.values[~np.eye(len(correlation_matrix), dtype=bool)].max():.3f}")

# Identifikasi kandidat untuk redundansi
moderate_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_val = correlation_matrix.iloc[i, j]
        if 0.5 <= abs(corr_val) <= high_correlation_threshold:
            var1 = correlation_matrix.columns[i]
            var2 = correlation_matrix.columns[j]
            moderate_corr_pairs.append((var1, var2, corr_val))

print(f"‚úì Korelasi moderate (0.5 ‚â§ |r| ‚â§ {high_correlation_threshold}): {len(moderate_corr_pairs)} pasangan")
if moderate_corr_pairs:
    print("  Kandidat untuk analisis lebih lanjut:")
    for var1, var2, corr in moderate_corr_pairs[:5]:  # Show top 5
        print(f"    {var1} ‚Üî {var2}: r = {corr:.3f}")
````

## 2. Covariance Analysis - Memahami Hubungan Linear

````python
# Covariance Analysis
print("=== COVARIANCE ANALYSIS ===")

# Hitung covariance matrix
covariance_matrix = df_integration[covid_numeric_cols].cov()

print("Covariance Matrix (excerpt - top 5x5):")
print(covariance_matrix.iloc[:5, :5])

# Normalisasi covariance untuk interpretasi yang lebih baik
# Standardisasi data untuk covariance analysis
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
covid_data_scaled = scaler.fit_transform(df_integration[covid_numeric_cols])
covid_df_scaled = pd.DataFrame(covid_data_scaled, columns=covid_numeric_cols)

# Covariance dari data yang sudah di-standardisasi (sama dengan correlation)
covariance_scaled = covid_df_scaled.cov()

# Visualisasi Covariance
fig, axes = plt.subplots(2, 2, figsize=(20, 16))

# 1. Covariance Matrix Original (log scale karena range yang besar)
covariance_log = np.log1p(np.abs(covariance_matrix))
im1 = axes[0,0].imshow(covariance_log, cmap='viridis', aspect='auto')
axes[0,0].set_title('Covariance Matrix (Log Scale)')
axes[0,0].set_xticks(range(len(covid_numeric_cols)))
axes[0,0].set_yticks(range(len(covid_numeric_cols)))
axes[0,0].set_xticklabels(covid_numeric_cols, rotation=45)
axes[0,0].set_yticklabels(covid_numeric_cols)
plt.colorbar(im1, ax=axes[0,0])

# 2. Covariance dari data standardized
sns.heatmap(covariance_scaled, annot=True, cmap='RdBu_r', center=0,
            square=True, ax=axes[0,1], cbar_kws={"shrink": .8})
axes[0,1].set_title('Covariance Matrix - Standardized Data')

# 3. Eigenvalue decomposition untuk Principal Component Analysis preview
eigenvalues, eigenvectors = np.linalg.eigh(covariance_scaled)
eigenvalues = eigenvalues[::-1]  # Sort descending
explained_variance_ratio = eigenvalues / eigenvalues.sum()

axes[1,0].bar(range(len(eigenvalues)), explained_variance_ratio)
axes[1,0].set_title('Explained Variance Ratio (Principal Components)')
axes[1,0].set_xlabel('Principal Component')
axes[1,0].set_ylabel('Explained Variance Ratio')
axes[1,0].set_xticks(range(len(eigenvalues)))

# Add cumulative variance
cumsum_variance = np.cumsum(explained_variance_ratio)
ax_twin = axes[1,0].twinx()
ax_twin.plot(range(len(eigenvalues)), cumsum_variance, 'r-o', alpha=0.7)
ax_twin.set_ylabel('Cumulative Variance Explained', color='red')

# 4. Variance Analysis per variabel
variances = df_integration[covid_numeric_cols].var().sort_values(ascending=False)
axes[1,1].barh(range(len(variances)), variances.values)
axes[1,1].set_yticks(range(len(variances)))
axes[1,1].set_yticklabels(variances.index)
axes[1,1].set_title('Variance per Variable')
axes[1,1].set_xlabel('Variance')

plt.tight_layout()
plt.show()

# Analisis Eigenvalues untuk dimensionality reduction insight
print("\n=== EIGENVALUE ANALYSIS ===")
print("Eigenvalues (descending):")
for i, (eigenval, var_ratio) in enumerate(zip(eigenvalues, explained_variance_ratio)):
    print(f"  PC{i+1}: Œª={eigenval:.3f}, Explained Variance={var_ratio:.1%}")

print(f"\nCumulative variance explained:")
for i, cum_var in enumerate(cumsum_variance):
    if cum_var >= 0.8 and i <= len(cumsum_variance)//2:  # Show up to 80% or half components
        print(f"  First {i+1} components: {cum_var:.1%}")

# Identifikasi variabel dengan covariance tinggi
print("\n=== HIGH COVARIANCE PAIRS ===")
high_cov_threshold = np.percentile(np.abs(covariance_scaled.values[~np.eye(len(covariance_scaled), dtype=bool)]), 75)
print(f"Threshold covariance tinggi (75th percentile): {high_cov_threshold:.3f}")

high_cov_pairs = []
for i in range(len(covariance_scaled.columns)):
    for j in range(i+1, len(covariance_scaled.columns)):
        cov_val = covariance_scaled.iloc[i, j]
        if abs(cov_val) > high_cov_threshold:
            var1 = covariance_scaled.columns[i]
            var2 = covariance_scaled.columns[j]
            high_cov_pairs.append((var1, var2, cov_val))

print(f"Pasangan dengan covariance tinggi:")
for var1, var2, cov in high_cov_pairs:
    print(f"  {var1:20s} ‚Üî {var2:20s}: cov = {cov:6.3f}")

# Summary
print(f"\n=== SUMMARY COVARIANCE ANALYSIS ===")
print(f"‚úì Eigenvalue terbesar: {eigenvalues[0]:.3f}")
print(f"‚úì Variance explained oleh PC1: {explained_variance_ratio[0]:.1%}")
print(f"‚úì Jumlah komponen untuk 80% variance: {np.sum(cumsum_variance <= 0.8) + 1}")
print(f"‚úì Variabel dengan variance tertinggi: {variances.index[0]} ({variances.iloc[0]:.1f})")
print(f"‚úì Pasangan covariance tinggi: {len(high_cov_pairs)}")
````

## 3. Penanganan Redundansi Data berdasarkan Analisis

````python
# Penanganan Redundansi Data
print("=== STRATEGI PENANGANAN REDUNDANSI DATA ===")

# Berdasarkan hasil correlation dan covariance analysis
print("Berdasarkan analisis correlation dan covariance, ditemukan:")
print("1. Korelasi sangat tinggi (>0.95) antara beberapa variabel ODP/PDP")
print("2. Variabel dengan variance sangat tinggi: odp, selesai_pemantauan, pdp")
print("3. Principal Component 1 menjelaskan ~80% variance")

# Strategi 1: Identifikasi Cluster Variabel Redundan
redundant_clusters = {
    'ODP_Group': ['odp', 'proses_pemantauan', 'selesai_pemantauan'],
    'PDP_Group': ['pdp', 'masih_dirawat', 'pulang_dan_sehat'],
    'Confirmed_Group': ['positif', 'dirawat', 'sembuh', 'meninggal', 'self_isolation']
}

print(f"\n=== ANALISIS REDUNDANSI PER CLUSTER ===")
for cluster_name, variables in redundant_clusters.items():
    print(f"\n{cluster_name}:")
    cluster_corr = correlation_matrix.loc[variables, variables]
    
    # Cari pasangan dengan korelasi tertinggi dalam cluster
    max_corr = 0
    max_pair = None
    for i in range(len(variables)):
        for j in range(i+1, len(variables)):
            corr_val = cluster_corr.iloc[i, j]
            if abs(corr_val) > max_corr:
                max_corr = abs(corr_val)
                max_pair = (variables[i], variables[j])
    
    print(f"  Max correlation: {max_pair[0]} ‚Üî {max_pair[1]} = {max_corr:.3f}")
    
    # Analisis informasi yang redundan vs unik
    if max_corr > 0.9:
        print(f"  üî¥ HIGH REDUNDANCY - Kandidat untuk consolidation")
    elif max_corr > 0.7:
        print(f"  üü° MODERATE REDUNDANCY - Perlu evaluasi domain knowledge")
    else:
        print(f"  üü¢ LOW REDUNDANCY - Keep all variables")

# Strategi 2: Pendekatan Domain-Informed untuk COVID-19 Data
print(f"\n=== PENDEKATAN DOMAIN-INFORMED REDUNDANCY HANDLING ===")

# Buat dataset integrated dengan pendekatan konservatif (medical data)
df_integrated = df_integration.copy()

# Untuk variabel ODP - buat composite score instead of dropping
print("\n1. ODP Group - Composite Score Creation:")
print("   Alasan: odp = proses_pemantauan + selesai_pemantauan (by definition)")

# Validasi konsistensi ODP
odp_consistency = df_integrated['odp'] - (df_integrated['proses_pemantauan'] + df_integrated['selesai_pemantauan'])
inconsistent_odp = (odp_consistency != 0).sum()
print(f"   Inkonsistensi ODP: {inconsistent_odp} records ({inconsistent_odp/len(df_integrated)*100:.1f}%)")

if inconsistent_odp == 0:
    print("   ‚úÖ Data ODP konsisten - bisa simplifikasi")
    # Buat ODP rate (completion rate)
    df_integrated['odp_completion_rate'] = np.where(
        df_integrated['odp'] > 0, 
        df_integrated['selesai_pemantauan'] / df_integrated['odp'], 
        0
    )
    print("   ‚úÖ Created: odp_completion_rate (selesai/total)")
else:
    print("   ‚ö†Ô∏è  Data ODP tidak konsisten - pertahankan semua")

# Untuk variabel PDP - evaluasi medical significance
print("\n2. PDP Group - Medical Significance Preservation:")
print("   Alasan: Setiap variabel memiliki makna klinis yang berbeda")

# Buat PDP outcome rate
df_integrated['pdp_recovery_rate'] = np.where(
    df_integrated['pdp'] > 0,
    df_integrated['pulang_dan_sehat'] / df_integrated['pdp'],
    0
)
print("   ‚úÖ Created: pdp_recovery_rate (pulang_sehat/total)")

# Untuk variabel Confirmed - buat comprehensive metrics
print("\n3. Confirmed Group - Comprehensive Metrics:")

# Case Fatality Rate (CFR)
df_integrated['case_fatality_rate'] = np.where(
    df_integrated['positif'] > 0,
    df_integrated['meninggal'] / df_integrated['positif'],
    0
)

# Recovery Rate
df_integrated['recovery_rate'] = np.where(
    df_integrated['positif'] > 0,
    df_integrated['sembuh'] / df_integrated['positif'],
    0
)

# Active Case Rate
df_integrated['active_case_rate'] = np.where(
    df_integrated['positif'] > 0,
    (df_integrated['dirawat'] + df_integrated['self_isolation']) / df_integrated['positif'],
    0
)

print("   ‚úÖ Created: case_fatality_rate, recovery_rate, active_case_rate")

# Strategi 3: Dimensionality Reduction Insight (tapi tidak applied untuk medical data)
print(f"\n=== DIMENSIONALITY REDUCTION INSIGHT ===")
print("Berdasarkan PCA analysis:")
print(f"- PC1 menjelaskan {explained_variance_ratio[0]:.1%} variance")
print(f"- {np.sum(cumsum_variance <= 0.95) + 1} komponen menjelaskan 95% variance")
print(f"- Rekomendasi: Pertahankan semua variabel original untuk interpretability")

# Summary variabel baru yang dibuat
new_features = ['odp_completion_rate', 'pdp_recovery_rate', 'case_fatality_rate', 
                'recovery_rate', 'active_case_rate']

print(f"\n=== SUMMARY REDUNDANCY HANDLING ===")
print(f"‚úÖ Pendekatan: Domain-informed feature engineering")
print(f"‚úÖ Original variables: {len(covid_numeric_cols)} (preserved)")
print(f"‚úÖ New engineered features: {len(new_features)}")
print(f"‚úÖ Total features after integration: {len(covid_numeric_cols) + len(new_features)}")
print(f"‚úÖ Redundansi ditangani dengan: Composite metrics, tidak dengan dropping")

# Tampilkan preview dataset integrated
print(f"\n=== PREVIEW NEW FEATURES ===")
print("New features statistics:")
for feature in new_features:
    if feature in df_integrated.columns:
        stats = df_integrated[feature].describe()
        print(f"{feature:20s}: Mean={stats['mean']:.3f}, Std={stats['std']:.3f}, Max={stats['max']:.3f}")

# Cek apakah ada nilai yang invalid (nan, inf)
invalid_values = df_integrated[new_features].isnull().sum().sum()
print(f"\nInvalid values in new features: {invalid_values}")

if invalid_values > 0:
    print("Cleaning invalid values...")
    for feature in new_features:
        if feature in df_integrated.columns:
            df_integrated[feature] = df_integrated[feature].fillna(0)
            df_integrated[feature] = df_integrated[feature].replace([np.inf, -np.inf], 0)
    print("‚úÖ Invalid values cleaned")
````

## 4. Advanced Feature Engineering untuk Dataset COVID-19

````python
# Advanced Feature Engineering
print("=== ADVANCED FEATURE ENGINEERING ===")

# 1. Epidemiological Risk Indicators
print("1. Creating Epidemiological Risk Indicators...")

# Total Surveillance Load per Kelurahan
df_integrated['total_surveillance'] = (df_integrated['odp'] + df_integrated['pdp'] + df_integrated['positif'])

# Epidemiological Pyramid Ratios
df_integrated['suspected_to_probable_ratio'] = np.where(
    df_integrated['pdp'] > 0,
    df_integrated['odp'] / df_integrated['pdp'],
    0
)

df_integrated['probable_to_confirmed_ratio'] = np.where(
    df_integrated['positif'] > 0,
    df_integrated['pdp'] / df_integrated['positif'],
    0
)

# 2. Severity and Outcome Indicators
print("2. Creating Severity and Outcome Indicators...")

# Hospitalization Rate
df_integrated['hospitalization_rate'] = np.where(
    df_integrated['positif'] > 0,
    df_integrated['dirawat'] / df_integrated['positif'],
    0
)

# Critical Care Indicator (high hospitalization + fatality)
df_integrated['severity_score'] = (
    0.3 * df_integrated['hospitalization_rate'] +
    0.7 * df_integrated['case_fatality_rate']
)

# 3. Geographic/Demographic Proxies
print("3. Creating Geographic/Administrative Features...")

# Create encoded features for administrative levels
from sklearn.preprocessing import LabelEncoder

# Encode Kecamatan (district) for potential clustering
le_kecamatan = LabelEncoder()
df_integrated['kecamatan_encoded'] = le_kecamatan.fit_transform(df_integrated['nama_kecamatan'])

# Create kecamatan-level aggregated features (for comparison)
kecamatan_stats = df_integrated.groupby('nama_kecamatan').agg({
    'positif': ['mean', 'sum'],
    'total_surveillance': ['mean', 'sum'],
    'case_fatality_rate': 'mean',
    'recovery_rate': 'mean'
}).round(3)

# Flatten column names
kecamatan_stats.columns = ['_'.join(col).strip() for col in kecamatan_stats.columns.values]

# Merge back kecamatan-level statistics
kecamatan_stats_dict = {}
for idx, row in kecamatan_stats.iterrows():
    kecamatan_stats_dict[idx] = {
        'kecamatan_avg_positif': row['positif_mean'],
        'kecamatan_total_positif': row['positif_sum'],
        'kecamatan_avg_cfr': row['case_fatality_rate_mean'],
        'kecamatan_avg_recovery': row['recovery_rate_mean']
    }

# Add kecamatan-level features to main dataset
for col in ['kecamatan_avg_positif', 'kecamatan_total_positif', 'kecamatan_avg_cfr', 'kecamatan_avg_recovery']:
    df_integrated[col] = df_integrated['nama_kecamatan'].map(
        lambda x: kecamatan_stats_dict[x][col] if x in kecamatan_stats_dict else 0
    )

# 4. Temporal Features (from periode_data)
print("4. Creating Temporal Features...")

# Extract temporal information
df_integrated['year'] = df_integrated['periode_data'] // 100
df_integrated['month'] = df_integrated['periode_data'] % 100

# 5. Composite Risk Indices
print("5. Creating Composite Risk Indices...")

# Overall Risk Index (weighted combination)
risk_components = {
    'surveillance_load': df_integrated['total_surveillance'] / df_integrated['total_surveillance'].max(),
    'severity': df_integrated['severity_score'],
    'transmission_risk': df_integrated['suspected_to_probable_ratio'] / (df_integrated['suspected_to_probable_ratio'].max() + 1e-6)
}

# Normalize components to 0-1 scale
for component, values in risk_components.items():
    if values.max() > 0:
        risk_components[component] = values / values.max()

# Combined Risk Index
df_integrated['composite_risk_index'] = (
    0.4 * risk_components['surveillance_load'] +
    0.4 * risk_components['severity'] +
    0.2 * risk_components['transmission_risk']
)

# 6. Data Quality and Completeness Features
print("6. Creating Data Quality Features...")

# Data completeness score
essential_fields = ['odp', 'pdp', 'positif', 'sembuh', 'meninggal']
df_integrated['data_completeness'] = df_integrated[essential_fields].notna().sum(axis=1) / len(essential_fields)

# Data consistency flags
df_integrated['data_consistency_flag'] = 0

# Flag inconsistent records (where totals don't add up)
total_check = df_integrated['positif'] - (df_integrated['dirawat'] + df_integrated['sembuh'] + 
                                        df_integrated['meninggal'] + df_integrated['self_isolation'])
df_integrated.loc[total_check != 0, 'data_consistency_flag'] = 1

# Summary of new features
engineered_features = [
    'total_surveillance', 'suspected_to_probable_ratio', 'probable_to_confirmed_ratio',
    'hospitalization_rate', 'severity_score', 'kecamatan_encoded',
    'kecamatan_avg_positif', 'kecamatan_total_positif', 'kecamatan_avg_cfr', 'kecamatan_avg_recovery',
    'year', 'month', 'composite_risk_index', 'data_completeness', 'data_consistency_flag'
]

all_new_features = new_features + engineered_features

print(f"\n=== SUMMARY FEATURE ENGINEERING ===")
print(f"‚úÖ Original COVID-19 features: {len(covid_numeric_cols)}")
print(f"‚úÖ Basic engineered features: {len(new_features)}")
print(f"‚úÖ Advanced engineered features: {len(engineered_features)}")
print(f"‚úÖ Total new features created: {len(all_new_features)}")
print(f"‚úÖ Final dataset shape: {df_integrated.shape}")

print(f"\nNew features categories:")
print(f"  - Rate/Ratio features: 8 (completion, recovery, fatality, etc.)")
print(f"  - Geographic features: 5 (kecamatan encoding & aggregates)")
print(f"  - Risk indices: 3 (severity, composite risk, etc.)")
print(f"  - Temporal features: 2 (year, month)")
print(f"  - Quality features: 2 (completeness, consistency)")

# Clean any remaining invalid values
print(f"\nCleaning invalid values...")
for feature in all_new_features:
    if feature in df_integrated.columns:
        # Replace inf and -inf with 0
        df_integrated[feature] = df_integrated[feature].replace([np.inf, -np.inf], 0)
        # Fill NaN with 0
        df_integrated[feature] = df_integrated[feature].fillna(0)

# Final validation
final_nulls = df_integrated[all_new_features].isnull().sum().sum()
final_infs = np.isinf(df_integrated[all_new_features].select_dtypes(include=[np.number])).sum().sum()

print(f"‚úÖ Final validation: {final_nulls} nulls, {final_infs} infinities")
print(f"‚úÖ Feature engineering completed successfully!")
````

## 3.1 Dimensionality Reduction Analysis

````python
# ===================================================================
# 3.1 DIMENSIONALITY REDUCTION ANALYSIS
# ===================================================================

print("=== DIMENSIONALITY REDUCTION ANALYSIS ===")

# Persiapkan data numerik yang sudah di-scale
scaler_reduction = StandardScaler()
X_scaled = scaler_reduction.fit_transform(df_for_reduction)

print(f"üìä Data Preparation:")
print(f"  Original dimensions: {df_for_reduction.shape}")
print(f"  Scaled data shape: {X_scaled.shape}")

# --- 1. PRINCIPAL COMPONENT ANALYSIS (PCA) ---
print(f"\n1Ô∏è‚É£ PRINCIPAL COMPONENT ANALYSIS (PCA)")

# PCA untuk menentukan optimal number of components
pca_full = PCA()
X_pca_full = pca_full.fit_transform(X_scaled)

# Analisis explained variance
explained_var_ratio = pca_full.explained_variance_ratio_
cumulative_var_ratio = np.cumsum(explained_var_ratio)

print(f"  Total components: {len(explained_var_ratio)}")
print(f"  Variance explained by top 5 PCs: {explained_var_ratio[:5]}")

# Tentukan optimal number of components (95% variance)
n_components_95 = np.argmax(cumulative_var_ratio >= 0.95) + 1
n_components_90 = np.argmax(cumulative_var_ratio >= 0.90) + 1
n_components_85 = np.argmax(cumulative_var_ratio >= 0.85) + 1

print(f"  Components for 95% variance: {n_components_95}")
print(f"  Components for 90% variance: {n_components_90}")
print(f"  Components for 85% variance: {n_components_85}")

# PCA dengan different component numbers
pca_configs = {
    'PCA_95': {'n_components': n_components_95, 'variance_threshold': 0.95},
    'PCA_90': {'n_components': n_components_90, 'variance_threshold': 0.90},
    'PCA_85': {'n_components': n_components_85, 'variance_threshold': 0.85},
    'PCA_10': {'n_components': 10, 'variance_threshold': None},
    'PCA_5': {'n_components': 5, 'variance_threshold': None}
}

pca_results = {}
for config_name, config in pca_configs.items():
    pca = PCA(n_components=config['n_components'])
    X_reduced = pca.fit_transform(X_scaled)
    
    var_explained = np.sum(pca.explained_variance_ratio_)
    
    pca_results[config_name] = {
        'transformer': pca,
        'data_reduced': X_reduced,
        'n_components': config['n_components'],
        'variance_explained': var_explained,
        'reduction_ratio': config['n_components'] / X_scaled.shape[1],
        'feature_importance': pca.components_
    }
    
    print(f"  {config_name}: {config['n_components']} components, {var_explained:.3f} variance")

# --- 2. t-SNE ANALYSIS ---
print(f"\n2Ô∏è‚É£ t-SNE ANALYSIS (Non-linear Dimensionality Reduction)")

# t-SNE dengan multiple perplexity values
tsne_configs = [
    {'n_components': 2, 'perplexity': 30, 'name': 'tSNE_2D_p30'},
    {'n_components': 2, 'perplexity': 50, 'name': 'tSNE_2D_p50'},
    {'n_components': 3, 'perplexity': 30, 'name': 'tSNE_3D_p30'}
]

tsne_results = {}
for config in tsne_configs:
    print(f"  Computing {config['name']}...")
    
    # Gunakan PCA preprocessing untuk t-SNE (recommended)
    X_pca_prep = pca_results['PCA_10']['data_reduced']  # Use 10 PCA components as input
    
    tsne = TSNE(
        n_components=config['n_components'],
        perplexity=config['perplexity'],
        random_state=42,
        n_iter=1000
    )
    
    X_tsne = tsne.fit_transform(X_pca_prep)
    
    tsne_results[config['name']] = {
        'transformer': tsne,
        'data_reduced': X_tsne,
        'n_components': config['n_components'],
        'perplexity': config['perplexity'],
        'kl_divergence': tsne.kl_divergence_
    }
    
    print(f"    {config['name']}: KL divergence = {tsne.kl_divergence_:.3f}")

# --- 3. FACTOR ANALYSIS ---
print(f"\n3Ô∏è‚É£ FACTOR ANALYSIS")

# Factor Analysis dengan different number of factors
factor_configs = [3, 5, 8, 10, 15]
factor_results = {}

for n_factors in factor_configs:
    fa = FactorAnalysis(n_components=n_factors, random_state=42)
    X_factors = fa.fit_transform(X_scaled)
    
    # Calculate likelihood score
    log_likelihood = fa.score(X_scaled)
    
    factor_results[f'FA_{n_factors}'] = {
        'transformer': fa,
        'data_reduced': X_factors,
        'n_components': n_factors,
        'log_likelihood': log_likelihood,
        'loadings': fa.components_,
        'noise_variance': fa.noise_variance_
    }
    
    print(f"  FA_{n_factors}: {n_factors} factors, log-likelihood = {log_likelihood:.3f}")

# --- 4. VARIANCE THRESHOLD ANALYSIS ---
print(f"\n4Ô∏è‚É£ VARIANCE THRESHOLD FEATURE FILTERING")

variance_thresholds = [0.0, 0.01, 0.05, 0.1, 0.2]
variance_results = {}

for threshold in variance_thresholds:
    selector = VarianceThreshold(threshold=threshold)
    X_variance_filtered = selector.fit_transform(X_scaled)
    
    selected_features = selector.get_support()
    n_selected = np.sum(selected_features)
    
    variance_results[f'VarThresh_{threshold}'] = {
        'transformer': selector,
        'data_reduced': X_variance_filtered,
        'n_features_selected': n_selected,
        'feature_mask': selected_features,
        'reduction_ratio': n_selected / X_scaled.shape[1],
        'threshold': threshold
    }
    
    print(f"  Threshold {threshold}: {n_selected}/{X_scaled.shape[1]} features retained ({n_selected/X_scaled.shape[1]:.2%})")

print(f"\n‚úÖ Dimensionality Reduction Analysis completed!")
print(f"  PCA configurations: {len(pca_results)}")
print(f"  t-SNE configurations: {len(tsne_results)}")
print(f"  Factor Analysis configurations: {len(factor_results)}")
print(f"  Variance Threshold configurations: {len(variance_results)}")

# Summary of reduction methods
reduction_summary = {
    'Original_Features': X_scaled.shape[1],
    'PCA_95%': pca_results['PCA_95']['n_components'],
    'PCA_90%': pca_results['PCA_90']['n_components'],
    'PCA_85%': pca_results['PCA_85']['n_components'],
    'FA_Best': min(factor_results.keys(), key=lambda x: -factor_results[x]['log_likelihood']),
    'VarThresh_Best': max(variance_results.keys(), key=lambda x: variance_results[x]['n_features_selected'])
}

print(f"\nüìä Reduction Summary:")
for method, value in reduction_summary.items():
    if method.startswith('Original'):
        print(f"  {method}: {value} features")
    elif method.endswith('_Best'):
        if 'FA_' in str(value):
            best_fa = factor_results[value]
            print(f"  {method}: {value} ({best_fa['n_components']} factors, LL={best_fa['log_likelihood']:.3f})")
        else:
            best_var = variance_results[value] 
            print(f"  {method}: {value} ({best_var['n_features_selected']} features, thresh={best_var['threshold']})")
    else:
        print(f"  {method}: {value} components")
````

## 3.2 Feature Selection Techniques

````python
# ===================================================================
# 3.2 FEATURE SELECTION TECHNIQUES
# ===================================================================

print("=== FEATURE SELECTION TECHNIQUES ===")

# Definisikan target variables untuk supervised feature selection
# Untuk COVID-19 analysis, kita gunakan multiple targets
target_variables = ['positif', 'meninggal', 'sembuh']  # Key COVID outcomes

feature_selection_results = {}

for target in target_variables:
    print(f"\nüéØ TARGET: {target.upper()}")
    
    # Prepare target variable
    y = df_for_reduction[target].values
    X = df_for_reduction.drop(columns=target_variables).values  # Remove all targets from features
    feature_names = [col for col in df_for_reduction.columns if col not in target_variables]
    
    print(f"  Features for selection: {len(feature_names)}")
    print(f"  Target range: {y.min():.0f} to {y.max():.0f}")
    
    # --- 1. UNIVARIATE FEATURE SELECTION ---
    print(f"\n  1Ô∏è‚É£ UNIVARIATE FEATURE SELECTION")
    
    # F-score based selection (regression)
    k_values = [5, 10, 15, 20]
    univariate_results = {}
    
    for k in k_values:
        selector = SelectKBest(score_func=f_regression, k=k)
        X_selected = selector.fit_transform(X, y)
        
        # Get selected features info
        selected_mask = selector.get_support()
        selected_features = [feature_names[i] for i in range(len(feature_names)) if selected_mask[i]]
        scores = selector.scores_
        
        univariate_results[f'Univariate_k{k}'] = {
            'selector': selector,
            'selected_features': selected_features,
            'selected_mask': selected_mask,
            'scores': scores,
            'n_features': k,
            'data_selected': X_selected
        }
        
        avg_score = np.mean(scores[selected_mask])
        print(f"    K={k}: {len(selected_features)} features, avg F-score={avg_score:.3f}")
    
    # --- 2. RECURSIVE FEATURE ELIMINATION (RFE) ---
    print(f"\n  2Ô∏è‚É£ RECURSIVE FEATURE ELIMINATION (RFE)")
    
    # RFE with Random Forest
    rf_estimator = RandomForestRegressor(n_estimators=100, random_state=42)
    
    rfe_results = {}
    n_features_list = [5, 10, 15, 20]
    
    for n_features in n_features_list:
        rfe = RFE(estimator=rf_estimator, n_features_to_select=n_features)
        X_rfe = rfe.fit_transform(X, y)
        
        selected_mask = rfe.get_support()
        selected_features = [feature_names[i] for i in range(len(feature_names)) if selected_mask[i]]
        feature_ranking = rfe.ranking_
        
        rfe_results[f'RFE_n{n_features}'] = {
            'selector': rfe,
            'selected_features': selected_features,
            'selected_mask': selected_mask,
            'ranking': feature_ranking,
            'n_features': n_features,
            'data_selected': X_rfe
        }
        
        print(f"    N={n_features}: Selected {len(selected_features)} features")
        print(f"      Top features: {selected_features[:3]}...")
    
    # --- 3. MODEL-BASED FEATURE SELECTION ---
    print(f"\n  3Ô∏è‚É£ MODEL-BASED FEATURE SELECTION")
    
    # Lasso-based feature selection
    lasso_cv = LassoCV(cv=5, random_state=42, max_iter=1000)
    lasso_cv.fit(X, y)
    
    # SelectFromModel with Lasso
    lasso_selector = SelectFromModel(lasso_cv, prefit=True)
    X_lasso = lasso_selector.transform(X)
    
    lasso_mask = lasso_selector.get_support()
    lasso_features = [feature_names[i] for i in range(len(feature_names)) if lasso_mask[i]]
    
    # Random Forest feature importance
    rf_importance = RandomForestRegressor(n_estimators=100, random_state=42)
    rf_importance.fit(X, y)
    
    # SelectFromModel with RF
    rf_selector = SelectFromModel(rf_importance, prefit=True)
    X_rf = rf_selector.transform(X)
    
    rf_mask = rf_selector.get_support()
    rf_features = [feature_names[i] for i in range(len(feature_names)) if rf_mask[i]]
    
    model_based_results = {
        'Lasso_Selection': {
            'selector': lasso_selector,
            'selected_features': lasso_features,
            'selected_mask': lasso_mask,
            'n_features': len(lasso_features),
            'data_selected': X_lasso,
            'coefficients': lasso_cv.coef_,
            'alpha': lasso_cv.alpha_
        },
        'RandomForest_Selection': {
            'selector': rf_selector,
            'selected_features': rf_features,
            'selected_mask': rf_mask,
            'n_features': len(rf_features),
            'data_selected': X_rf,
            'feature_importances': rf_importance.feature_importances_
        }
    }
    
    print(f"    Lasso Selection: {len(lasso_features)} features (alpha={lasso_cv.alpha_:.4f})")
    print(f"      Selected: {lasso_features[:3] if len(lasso_features) >= 3 else lasso_features}...")
    
    print(f"    Random Forest Selection: {len(rf_features)} features")
    print(f"      Selected: {rf_features[:3] if len(rf_features) >= 3 else rf_features}...")
    
    # Store all results for this target
    feature_selection_results[target] = {
        'univariate': univariate_results,
        'rfe': rfe_results,
        'model_based': model_based_results,
        'target_stats': {
            'mean': np.mean(y),
            'std': np.std(y),
            'min': np.min(y),
            'max': np.max(y)
        }
    }

# --- 4. CONSENSUS FEATURE SELECTION ---
print(f"\n4Ô∏è‚É£ CONSENSUS FEATURE SELECTION")
print(f"Finding features commonly selected across different methods...")

# Analyze feature selection consensus across targets and methods
consensus_analysis = {}

for target in target_variables:
    target_consensus = {}
    
    # Collect all selected features for this target
    all_selected_features = []
    method_names = []
    
    # From univariate (use k=10 as representative)
    univariate_features = feature_selection_results[target]['univariate']['Univariate_k10']['selected_features']
    all_selected_features.extend(univariate_features)
    method_names.extend([f'Univariate_{target}'] * len(univariate_features))
    
    # From RFE (use n=10 as representative)
    rfe_features = feature_selection_results[target]['rfe']['RFE_n10']['selected_features']
    all_selected_features.extend(rfe_features)
    method_names.extend([f'RFE_{target}'] * len(rfe_features))
    
    # From Lasso
    lasso_features = feature_selection_results[target]['model_based']['Lasso_Selection']['selected_features']
    all_selected_features.extend(lasso_features)
    method_names.extend([f'Lasso_{target}'] * len(lasso_features))
    
    # From Random Forest
    rf_features = feature_selection_results[target]['model_based']['RandomForest_Selection']['selected_features']
    all_selected_features.extend(rf_features)
    method_names.extend([f'RF_{target}'] * len(rf_features))
    
    # Count frequency of each feature
    feature_counts = {}
    for feature in all_selected_features:
        feature_counts[feature] = feature_counts.get(feature, 0) + 1
    
    # Sort by frequency
    sorted_features = sorted(feature_counts.items(), key=lambda x: x[1], reverse=True)
    
    target_consensus = {
        'feature_frequency': feature_counts,
        'sorted_features': sorted_features,
        'top_10': [f[0] for f in sorted_features[:10]],
        'highly_selected': [f[0] for f in sorted_features if f[1] >= 3]  # Selected by 3+ methods
    }
    
    consensus_analysis[target] = target_consensus
    
    print(f"\n  Target {target.upper()} - Top consensus features:")
    for i, (feature, count) in enumerate(sorted_features[:5]):
        print(f"    {i+1}. {feature}: selected by {count}/4 methods")

# Cross-target consensus (features important for multiple COVID outcomes)
print(f"\n5Ô∏è‚É£ CROSS-TARGET CONSENSUS")
cross_target_features = {}

for target in target_variables:
    top_features = consensus_analysis[target]['top_10']
    for feature in top_features:
        cross_target_features[feature] = cross_target_features.get(feature, 0) + 1

cross_target_sorted = sorted(cross_target_features.items(), key=lambda x: x[1], reverse=True)
universal_features = [f[0] for f in cross_target_sorted if f[1] >= 2]  # Important for 2+ targets

print(f"Universal features (important for multiple COVID outcomes):")
for i, (feature, count) in enumerate(cross_target_sorted[:10]):
    print(f"  {i+1}. {feature}: important for {count}/3 targets")

print(f"\n‚úÖ Feature Selection Analysis completed!")
print(f"  Targets analyzed: {len(target_variables)}")
print(f"  Methods per target: 4 (Univariate, RFE, Lasso, Random Forest)")
print(f"  Universal features identified: {len(universal_features)}")
print(f"  Total feature selection configurations: {sum([len(fs['univariate']) + len(fs['rfe']) + len(fs['model_based']) for fs in feature_selection_results.values()])}")

# Store consensus results
feature_selection_results['consensus'] = {
    'cross_target_consensus': cross_target_features,
    'universal_features': universal_features,
    'target_specific_consensus': consensus_analysis
}
````

## 3.3 Statistical Data Reduction

````python
# ===================================================================
# 3.3 STATISTICAL DATA REDUCTION
# ===================================================================

print("=== STATISTICAL DATA REDUCTION ===")

# --- 1. CLUSTERING-BASED DATA REDUCTION ---
print(f"\n1Ô∏è‚É£ CLUSTERING-BASED DATA REDUCTION")

# K-Means clustering untuk identify representative samples
clustering_results = {}

# Test different numbers of clusters
n_clusters_list = [3, 5, 8, 10, 15, 20]

for n_clusters in n_clusters_list:
    print(f"\n  Testing {n_clusters} clusters...")
    
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_scaled)
    
    # Calculate cluster statistics
    cluster_centers = kmeans.cluster_centers_
    inertia = kmeans.inertia_
    
    # For each cluster, find representative samples
    cluster_representatives = []
    cluster_info = []
    
    for cluster_id in range(n_clusters):
        cluster_mask = cluster_labels == cluster_id
        cluster_data = X_scaled[cluster_mask]
        cluster_size = np.sum(cluster_mask)
        
        if cluster_size > 0:
            # Find closest sample to cluster center
            distances = np.linalg.norm(cluster_data - cluster_centers[cluster_id], axis=1)
            closest_idx = np.argmin(distances)
            
            # Get original index
            original_indices = np.where(cluster_mask)[0]
            representative_idx = original_indices[closest_idx]
            
            cluster_representatives.append(representative_idx)
            cluster_info.append({
                'cluster_id': cluster_id,
                'size': cluster_size,
                'representative_idx': representative_idx,
                'avg_distance_to_center': np.mean(distances)
            })
    
    # Create reduced dataset with representatives only
    representative_data = df_for_reduction.iloc[cluster_representatives].copy()
    representative_data['cluster_id'] = range(len(cluster_representatives))
    
    clustering_results[f'KMeans_{n_clusters}'] = {
        'model': kmeans,
        'n_clusters': n_clusters,
        'labels': cluster_labels,
        'centers': cluster_centers,
        'inertia': inertia,
        'representatives': cluster_representatives,
        'cluster_info': cluster_info,
        'reduced_data': representative_data,
        'reduction_ratio': len(cluster_representatives) / len(df_for_reduction)
    }
    
    print(f"    Inertia: {inertia:.3f}")
    print(f"    Representatives: {len(cluster_representatives)} samples ({len(cluster_representatives)/len(df_for_reduction):.2%} of original)")

# --- 2. SAMPLING-BASED REDUCTION ---
print(f"\n2Ô∏è‚É£ SAMPLING-BASED REDUCTION")

sampling_results = {}

# Stratified sampling based on target variables
print(f"  Original dataset size: {len(df_for_reduction)}")

# Create strata based on COVID severity (using positif cases)
positif_values = df_for_reduction['positif'].values
positif_quartiles = np.percentile(positif_values, [25, 50, 75])

def get_stratum(value):
    if value <= positif_quartiles[0]:
        return 'Low'
    elif value <= positif_quartiles[1]:
        return 'Medium-Low' 
    elif value <= positif_quartiles[2]:
        return 'Medium-High'
    else:
        return 'High'

df_for_reduction['severity_stratum'] = df_for_reduction['positif'].apply(get_stratum)

# Different sampling ratios
sampling_ratios = [0.3, 0.5, 0.7, 0.8]

for ratio in sampling_ratios:
    # Stratified sampling
    sampled_indices = []
    
    for stratum in ['Low', 'Medium-Low', 'Medium-High', 'High']:
        stratum_data = df_for_reduction[df_for_reduction['severity_stratum'] == stratum]
        stratum_size = len(stratum_data)
        sample_size = max(1, int(stratum_size * ratio))  # At least 1 sample per stratum
        
        if stratum_size > 0:
            sampled_stratum = stratum_data.sample(n=sample_size, random_state=42)
            sampled_indices.extend(sampled_stratum.index.tolist())
    
    sampled_data = df_for_reduction.loc[sampled_indices].copy()
    
    # Calculate sampling statistics
    original_stratum_counts = df_for_reduction['severity_stratum'].value_counts()
    sampled_stratum_counts = sampled_data['severity_stratum'].value_counts()
    
    sampling_results[f'Stratified_{int(ratio*100)}pct'] = {
        'ratio': ratio,
        'sampled_indices': sampled_indices,
        'sampled_data': sampled_data,
        'original_size': len(df_for_reduction),
        'sampled_size': len(sampled_data),
        'actual_ratio': len(sampled_data) / len(df_for_reduction),
        'stratum_preservation': sampled_stratum_counts / original_stratum_counts
    }
    
    print(f"  Stratified {int(ratio*100)}%: {len(sampled_data)} samples ({len(sampled_data)/len(df_for_reduction):.2%})")

# Random sampling for comparison
for ratio in sampling_ratios:
    sample_size = int(len(df_for_reduction) * ratio)
    random_sampled = df_for_reduction.sample(n=sample_size, random_state=42)
    
    sampling_results[f'Random_{int(ratio*100)}pct'] = {
        'ratio': ratio,
        'sampled_indices': random_sampled.index.tolist(),
        'sampled_data': random_sampled,
        'original_size': len(df_for_reduction),
        'sampled_size': len(random_sampled),
        'actual_ratio': len(random_sampled) / len(df_for_reduction)
    }
    
    print(f"  Random {int(ratio*100)}%: {len(random_sampled)} samples")

# --- 3. BINNING/DISCRETIZATION ---
print(f"\n3Ô∏è‚É£ DISCRETIZATION & BINNING")

discretization_results = {}

# Identify continuous variables for binning
continuous_vars = ['positif', 'meninggal', 'sembuh', 'odp', 'pdp', 
                   'composite_risk_index', 'severity_score', 'hospitalization_rate']

bin_strategies = {
    'uniform': 'uniform',
    'quantile': 'quantile',
    'kmeans': 'kmeans'
}

n_bins_list = [3, 5, 8]

for strategy_name, strategy in bin_strategies.items():
    for n_bins in n_bins_list:
        print(f"  {strategy_name.title()} binning with {n_bins} bins...")
        
        discretizer = KBinsDiscretizer(n_bins=n_bins, encode='ordinal', strategy=strategy, random_state=42)
        
        # Apply to continuous variables
        continuous_data = df_for_reduction[continuous_vars].values
        binned_data = discretizer.fit_transform(continuous_data)
        
        # Create binned dataset
        binned_df = df_for_reduction.copy()
        for i, var in enumerate(continuous_vars):
            binned_df[f'{var}_binned'] = binned_data[:, i]
        
        # Calculate information loss
        original_unique_values = sum([df_for_reduction[var].nunique() for var in continuous_vars])
        binned_unique_values = sum([binned_df[f'{var}_binned'].nunique() for var in continuous_vars])
        
        compression_ratio = binned_unique_values / original_unique_values
        
        discretization_results[f'{strategy_name}_{n_bins}bins'] = {
            'discretizer': discretizer,
            'n_bins': n_bins,
            'strategy': strategy,
            'binned_data': binned_df,
            'continuous_vars': continuous_vars,
            'compression_ratio': compression_ratio,
            'original_unique_total': original_unique_values,
            'binned_unique_total': binned_unique_values,
            'bin_edges': [discretizer.bin_edges_[i] for i in range(len(continuous_vars))]
        }
        
        print(f"    Compression ratio: {compression_ratio:.3f} ({original_unique_values} ‚Üí {binned_unique_values} unique values)")

# --- 4. CORRELATION-BASED FEATURE GROUPING ---
print(f"\n4Ô∏è‚É£ CORRELATION-BASED FEATURE GROUPING")

# Group highly correlated features and select representatives
correlation_matrix = df_for_reduction.select_dtypes(include=[np.number]).corr()

correlation_threshold = 0.8
feature_groups = []
processed_features = set()

numeric_features = correlation_matrix.columns.tolist()

for i, feature1 in enumerate(numeric_features):
    if feature1 in processed_features:
        continue
    
    # Find features highly correlated with current feature
    high_corr_features = []
    for j, feature2 in enumerate(numeric_features):
        if i != j and abs(correlation_matrix.loc[feature1, feature2]) >= correlation_threshold:
            high_corr_features.append(feature2)
    
    if high_corr_features:
        # Create group
        group = [feature1] + high_corr_features
        feature_groups.append(group)
        
        # Mark as processed
        for f in group:
            processed_features.add(f)
    else:
        # Single feature group
        feature_groups.append([feature1])
        processed_features.add(feature1)

print(f"  Identified {len(feature_groups)} feature groups:")
for i, group in enumerate(feature_groups):
    if len(group) > 1:
        print(f"    Group {i+1}: {len(group)} features - {group}")

# For each group, select representative (highest variance or domain importance)
representative_features = []
for group in feature_groups:
    if len(group) == 1:
        representative_features.append(group[0])
    else:
        # Select feature with highest variance (most information)
        group_variances = {f: df_for_reduction[f].var() for f in group}
        representative = max(group_variances, key=group_variances.get)
        representative_features.append(representative)
        print(f"    Group representative: {representative} (variance: {group_variances[representative]:.3f})")

# Create correlation-reduced dataset
corr_reduced_data = df_for_reduction[representative_features].copy()

correlation_reduction_results = {
    'threshold': correlation_threshold,
    'feature_groups': feature_groups,
    'representative_features': representative_features,
    'reduced_data': corr_reduced_data,
    'original_features': len(numeric_features),
    'reduced_features': len(representative_features),
    'reduction_ratio': len(representative_features) / len(numeric_features)
}

print(f"  Correlation-based reduction: {len(numeric_features)} ‚Üí {len(representative_features)} features ({len(representative_features)/len(numeric_features):.2%})")

# Clean up temporary column
df_for_reduction.drop('severity_stratum', axis=1, inplace=True)

print(f"\n‚úÖ Statistical Data Reduction completed!")
print(f"  Clustering configurations: {len(clustering_results)}")
print(f"  Sampling configurations: {len(sampling_results)}")
print(f"  Discretization configurations: {len(discretization_results)}")
print(f"  Correlation-based reduction: {len(representative_features)} representative features")

# Summary of statistical reduction methods
statistical_summary = {
    'Original_Samples': len(df_for_reduction),
    'Best_Clustering': min(clustering_results.keys(), key=lambda x: clustering_results[x]['inertia']),
    'Best_Sampling_50pct': 'Stratified_50pct',
    'Best_Binning': 'quantile_5bins',
    'Correlation_Reduced_Features': len(representative_features)
}

print(f"\nüìä Statistical Reduction Summary:")
for method, value in statistical_summary.items():
    if method.startswith('Original'):
        print(f"  {method}: {value} samples")
    elif 'Clustering' in method:
        best_clustering = clustering_results[value]
        print(f"  {method}: {value} ({best_clustering['n_clusters']} clusters, {best_clustering['reduction_ratio']:.2%} retained)")
    elif 'Features' in method:
        print(f"  {method}: {value} features ({value}/{len(numeric_features)} = {value/len(numeric_features):.2%})")
    else:
        print(f"  {method}: {value}")

# Store results for later use
statistical_reduction_results = {
    'clustering': clustering_results,
    'sampling': sampling_results,
    'discretization': discretization_results,
    'correlation_reduction': correlation_reduction_results
}
````
